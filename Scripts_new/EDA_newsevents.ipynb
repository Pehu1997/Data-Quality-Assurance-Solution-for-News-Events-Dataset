```python
from urllib.parse import urlparse

def is_valid_url(url):
    try:
        result = urlparse(url)
        return all([result.scheme, result.netloc])  # checks if "http(s)" and domain exist
    except:
        return False

# Apply the function
df['valid_url'] = df['article_url'].apply(lambda x: is_valid_url(str(x)))

# Summary
print(df['valid_url'].value_counts())
```

```python
df['article_domain'] = df['article_url'].apply(lambda x: urlparse(str(x)).netloc if is_valid_url(str(x)) else None)

# Top 10 most common news domains
top_domains = df['article_domain'].value_counts().head(10)
print("Top 10 Domains:\n", top_domains)
```

```python
import requests

def check_url_status(url):
    try:
        response = requests.head(url, timeout=5)
        return response.status_code
    except:
        return None

# Apply to a small sample (first 10)
df_sample = df.head(10).copy()
df_sample['url_status'] = df_sample['article_url'].apply(check_url_status)
```
